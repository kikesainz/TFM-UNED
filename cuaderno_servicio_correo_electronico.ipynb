## 1. Historia y evolución del correo electrónico

El correo electrónico no apareció como un servicio aislado de Internet, sino que fue evolucionando junto con los primeros sistemas multiusuario y las redes de ordenadores.

| Año / etapa | Hito principal | Importancia |
|---|---|---|
| 1965 | Mensajes electrónicos en CTSS, en el MIT | Permitía dejar mensajes a otros usuarios del mismo sistema. |
| 1971 | Ray Tomlinson desarrolla el correo electrónico en ARPANET | Introduce el uso del símbolo `@` para separar usuario y dominio. |
| 1977 | Estandarización del formato de correo | Se definen campos como `To`, `From` y `Subject`. |
| 1982 | Creación de SMTP | Se establece un estándar para el envío de mensajes entre servidores. |
| 1984 | Introducción de POP | Permite descargar mensajes desde un servidor a un cliente. |
| Década de 1990 | Popularización con Internet | Aparecen servicios como AOL, Hotmail y Yahoo! Mail. |
| 2004 | Lanzamiento de Gmail | Impulsa el correo web con gran almacenamiento, búsqueda avanzada e integración con otros servicios. |
| 2000s-presente | Seguridad, movilidad e integración | SPF, DKIM, DMARC, acceso móvil, IA, calendarios y almacenamiento en la nube. |

## 3. Componentes básicos de un sistema de correo electrónico

Un sistema de correo electrónico está formado por varios elementos que trabajan de forma coordinada para enviar, recibir y almacenar mensajes.

Los tres componentes principales son:

| Componente | Nombre completo | Función principal | Ejemplos |
|---|---|---|---|
| **MUA** | Mail User Agent | Cliente usado por el usuario para redactar, enviar y leer correos. | Outlook, Thunderbird, Gmail, Apple Mail. |
| **MTA** | Mail Transfer Agent | Servidor que transporta mensajes entre servidores usando SMTP. | Postfix, Exim, Sendmail, Microsoft Exchange. |
| **MDA** | Mail Delivery Agent | Recibe el mensaje y lo almacena en el buzón del destinatario. | Dovecot, Courier, Procmail. |


### 3.1. MUA: Mail User Agent

El **MUA** es el cliente de correo utilizado por el usuario final.

Sus funciones principales son:

- Redactar y enviar mensajes.
- Leer, responder, reenviar y eliminar correos.
- Organizar mensajes mediante carpetas, etiquetas o filtros.
- Conectarse con servidores usando protocolos como SMTP, POP3 o IMAP.

Ejemplos:

- Clientes de escritorio: **Microsoft Outlook**, **Mozilla Thunderbird**.
- Aplicaciones móviles: **Gmail**, **Apple Mail**.
- Clientes web: **Gmail**, **Outlook.com**, **Yahoo! Mail**.


### 3.2. MTA: Mail Transfer Agent

El **MTA** es el servidor encargado de transportar correos electrónicos.

Sus funciones principales son:

- Recibir mensajes enviados por un MUA u otro MTA.
- Consultar DNS para localizar el servidor de correo del dominio destinatario.
- Transferir el mensaje al servidor correspondiente mediante SMTP.
- Gestionar errores de entrega, reintentos y mensajes de rebote.

Ejemplos de MTA:

- **Postfix**.
- **Exim**.
- **Sendmail**.
- **Microsoft Exchange Server**.


### 3.3. MDA: Mail Delivery Agent

El **MDA** es el agente encargado de entregar y almacenar el mensaje en el buzón del destinatario.

Sus funciones principales son:

- Recibir mensajes desde el MTA.
- Guardarlos en el buzón del usuario.
- Aplicar filtros o reglas de clasificación.
- Permitir que los clientes accedan al buzón mediante POP3 o IMAP.

Ejemplos de MDA o servidores de acceso al buzón:

- **Dovecot**.
- **Courier**.
- **Procmail**.


### 3.4. Relación entre MUA, MTA y MDA

El flujo básico de un correo electrónico puede resumirse así:

1. El usuario redacta un mensaje en su **MUA**.
2. El MUA envía el mensaje al **MTA** del remitente mediante **SMTP**.
3. El MTA consulta DNS, normalmente registros **MX**, para encontrar el servidor del destinatario.
4. El mensaje viaja entre servidores **MTA** usando SMTP.
5. El **MDA** del destinatario almacena el mensaje en su buzón.
6. El destinatario consulta el buzón desde su MUA mediante **POP3**, **IMAP** o webmail.

```text
Usuario remitente
      │
      ▼
MUA ──SMTP──▶ MTA remitente ──SMTP──▶ MTA destinatario ──▶ MDA ──IMAP/POP3──▶ MUA destinatario
```

Entre el MTA y el MDA pueden existir sistemas de seguridad como filtros antispam, antivirus o herramientas como **Apache SpamAssassin**.


````python
# Representación sencilla del flujo de correo electrónico
flujo = [
    "1. El usuario escribe el mensaje en el MUA",
    "2. El MUA envía el mensaje al MTA mediante SMTP",
    "3. El MTA consulta DNS para localizar el servidor del destinatario",
    "4. El mensaje viaja entre MTAs mediante SMTP",
    "5. El MDA almacena el mensaje en el buzón del destinatario",
    "6. El destinatario accede al buzón mediante IMAP, POP3 o webmail"
]

for paso in flujo:
    print(paso)
````

## 4. Estructura de un mensaje de correo electrónico

Un mensaje de correo electrónico está formado por tres partes principales:

1. **Cabeceras**.
2. **Cuerpo**.
3. **Adjuntos**.

Estos elementos permiten transmitir información de forma estructurada y compatible entre distintos sistemas de correo.


### 4.1. Cabeceras de un correo

Las cabeceras contienen información esencial sobre el mensaje y su recorrido.

Algunas cabeceras son visibles para el usuario, como el remitente o el asunto. Otras son técnicas y sirven para diagnosticar problemas, verificar autenticidad o analizar el trayecto seguido por el correo.

En muchos clientes de correo, para verlas hay que abrir el mensaje y seleccionar una opción como **Mostrar original**, **Ver código fuente** o **Ver cabeceras completas**.


### 4.2. Cabeceras habituales

| Cabecera | Función | Ejemplo |
|---|---|---|
| `From` | Remitente del mensaje. | `From: Juan Pérez <juan@ejemplo.com>` |
| `To` | Destinatario principal. | `To: María López <maria@ejemplo.com>` |
| `Subject` | Asunto del mensaje. | `Subject: Informe mensual` |
| `Date` | Fecha y hora de envío. | `Date: Tue, 20 Dec 2023 15:45:00 +0000` |
| `Cc` | Copia visible a otros destinatarios. | `Cc: equipo@ejemplo.com` |
| `Bcc` | Copia oculta. No aparece ante el resto. | No suele mostrarse al destinatario. |
| `Reply-To` | Dirección alternativa para responder. | `Reply-To: soporte@ejemplo.com` |
| `Message-ID` | Identificador único del mensaje. | `Message-ID: <abc123@ejemplo.com>` |


### 4.3. Cabeceras técnicas

| Cabecera | Función |
|---|---|
| `Received` | Registra el paso del mensaje por distintos servidores. Es clave para analizar el recorrido. |
| `Return-Path` | Dirección a la que se envían rebotes si el mensaje no puede entregarse. |
| `Content-Type` | Indica si el contenido es texto plano, HTML o multipart con adjuntos. |
| `MIME-Version` | Indica la versión del estándar MIME usada. |
| `Authentication-Results` | Resultado de validaciones como SPF, DKIM o DMARC. |
| `X-Headers` | Cabeceras personalizadas usadas por servidores o aplicaciones. |


````python
# Práctica: analizar cabeceras básicas de un mensaje con Python
from email import policy
from email.parser import Parser

mensaje = '''From: Juan Pérez <juan.perez@ejemplo.com>
To: María López <maria.lopez@destino.com>
Subject: Informe de ventas mensual
Date: Tue, 20 Dec 2023 15:45:00 +0000
Message-ID: <abc123@ejemplo.com>
Received: from mail.ejemplo.com by mx.destino.com with SMTP
Content-Type: text/plain; charset=utf-8

Hola María,
Te envío el informe de ventas mensual.
Un saludo.
'''

email = Parser(policy=policy.default).parsestr(mensaje)

for cabecera in ["From", "To", "Subject", "Date", "Message-ID", "Received", "Content-Type"]:
    print(f"{cabecera}: {email.get(cabecera)}")
````

### 4.4. Cuerpo del mensaje

El cuerpo es la parte principal del correo, donde se transmite el contenido.

Puede estar en:

- **Texto plano**: solo texto, sin formato.
- **HTML**: permite estilos, enlaces, imágenes incrustadas y estructura visual.
- **Multipart**: combina varias versiones del mensaje, por ejemplo texto plano y HTML.


### 4.5. Adjuntos y MIME

Los adjuntos permiten enviar documentos, imágenes, vídeos u otros archivos junto con el mensaje.

Para ello se usa **MIME** (*Multipurpose Internet Mail Extensions*), un estándar que permite:

- Dividir el correo en varias partes.
- Incluir texto, HTML y archivos adjuntos.
- Codificar contenido binario, normalmente en **Base64**, para transportarlo dentro de un mensaje de correo.

Cada adjunto incluye información como:

- Tipo de contenido.
- Nombre del archivo.
- Codificación utilizada.


````python
# Ejemplo: crear un mensaje multipart sencillo con texto y HTML
from email.message import EmailMessage

msg = EmailMessage()
msg["From"] = "profesor@ejemplo.com"
msg["To"] = "alumno@ejemplo.com"
msg["Subject"] = "Ejemplo de mensaje MIME"

msg.set_content("Este es el contenido en texto plano.")
msg.add_alternative('''
<html>
  <body>
    <p>Este es el contenido en <strong>HTML</strong>.</p>
  </body>
</html>
''', subtype="html")

print(msg)
````

## 5. Protocolos asociados al correo electrónico

Los protocolos de correo electrónico son conjuntos de reglas que permiten la comunicación entre clientes y servidores.

Los principales son:

| Protocolo | Uso principal | Puertos habituales |
|---|---|---|
| **SMTP** | Enviar correo desde cliente a servidor y entre servidores. | 25, 587, 465 |
| **POP3** | Descargar mensajes del servidor al cliente. | 110, 995 |
| **IMAP** | Acceder y sincronizar mensajes almacenados en el servidor. | 143, 993 |
| **HTTP/HTTPS** | Acceder al correo mediante navegador web, webmail. | 80, 443 |


### 5.1. SMTP: Simple Mail Transfer Protocol

**SMTP** es el protocolo estándar para el envío de correo electrónico.

Se utiliza para:

- Enviar mensajes desde un cliente de correo al servidor de salida.
- Transferir mensajes entre servidores de correo.

Puertos más habituales:

| Puerto | Uso |
|---|---|
| 25 | SMTP entre servidores. También se usaba para envío, pero suele estar restringido por SPAM. |
| 587 | Envío autenticado con STARTTLS. Es el puerto recomendado para clientes. |
| 465 | SMTP sobre SSL/TLS implícito. |


### 5.2. Flujo básico SMTP

Un envío SMTP básico sigue esta secuencia:

1. El cliente se conecta al servidor SMTP.
2. Se identifica con `HELO` o `EHLO`.
3. Indica el remitente con `MAIL FROM`.
4. Indica el destinatario con `RCPT TO`.
5. Envía el contenido del mensaje con `DATA`.
6. Finaliza la sesión con `QUIT`.

Ejemplo conceptual:

```text
EHLO cliente.ejemplo.com
MAIL FROM:<juan@ejemplo.com>
RCPT TO:<maria@destino.com>
DATA
Subject: Prueba SMTP

Hola, esto es una prueba.
.
QUIT
```


### 5.3. Comandos SMTP principales

| Comando | Función |
|---|---|
| `HELO` / `EHLO` | Identifica al cliente. `EHLO` permite extensiones modernas. |
| `MAIL FROM` | Indica el remitente. |
| `RCPT TO` | Indica el destinatario. |
| `DATA` | Inicia el envío de cabecera y cuerpo del mensaje. |
| `STARTTLS` | Solicita cifrar la conexión con TLS. |
| `AUTH` | Permite autenticación del cliente. |
| `QUIT` | Cierra la sesión. |


### 5.4. Seguridad en SMTP

SMTP nació sin mecanismos de seguridad fuertes. Por eso se han añadido mejoras:

- **STARTTLS**: cifra la conexión entre cliente y servidor o entre servidores.
- **SMTP AUTH**: exige autenticación para enviar correos.
- **SPF, DKIM y DMARC**: ayudan a evitar suplantación y falsificación de dominios.

Estas medidas son esenciales para reducir el uso no autorizado de servidores de correo y dificultar el SPAM y el phishing.


### 5.5. POP3: Post Office Protocol v3

**POP3** permite descargar mensajes desde el servidor al dispositivo del usuario.

Características principales:

- Descarga los mensajes al cliente.
- Puede borrar los mensajes del servidor tras descargarlos.
- Es menos adecuado cuando se consultan correos desde varios dispositivos.
- Es útil si se desea conservar el correo localmente.

Puertos habituales:

| Puerto | Uso |
|---|---|
| 110 | POP3 sin cifrado. |
| 995 | POP3 seguro con SSL/TLS. |


### 5.6. Flujo de trabajo POP3

1. El cliente se conecta al servidor POP3.
2. El usuario se autentica.
3. El cliente obtiene la lista de mensajes disponibles.
4. Los mensajes se descargan al dispositivo.
5. Opcionalmente, los mensajes se eliminan del servidor.
6. Se cierra la sesión.

POP3 puede ser adecuado para escenarios sencillos, pero presenta limitaciones cuando se trabaja desde varios dispositivos.


### 5.7. IMAP: Internet Message Access Protocol

**IMAP** permite acceder y gestionar correos almacenados en el servidor.

Características principales:

- Mantiene los mensajes en el servidor.
- Sincroniza carpetas, leídos/no leídos y organización entre dispositivos.
- Es ideal para usuarios que usan varios dispositivos.
- Permite consultar mensajes sin descargarlos completamente.

Puertos habituales:

| Puerto | Uso |
|---|---|
| 143 | IMAP sin cifrado. |
| 993 | IMAP seguro con SSL/TLS. |


### 5.8. Webmail mediante HTTP/HTTPS

El webmail permite acceder al correo desde un navegador.

Ejemplos:

- Gmail.
- Outlook.com.
- Yahoo! Mail.
- Proton Mail.
- Roundcube.
- Horde.

Normalmente el usuario accede mediante **HTTPS**, mientras que por debajo el servidor sigue utilizando protocolos de correo como SMTP, IMAP u otros mecanismos internos.


### 5.9. Comparativa POP3, IMAP y webmail

| Aspecto | POP3 | IMAP | Webmail |
|---|---|---|---|
| Ubicación principal del correo | Cliente local | Servidor | Servidor, accedido desde navegador |
| Sincronización multidispositivo | Limitada | Muy buena | Muy buena |
| Uso offline | Bueno si se ha descargado el correo | Depende de la configuración | Depende del navegador/app |
| Gestión de carpetas | Principalmente local | Sincronizada | Sincronizada |
| Caso típico | Un único equipo | Varios dispositivos | Acceso desde cualquier navegador |

## 7. Seguridad en el correo electrónico

El correo electrónico es un objetivo habitual de ataques porque se usa de forma masiva y suele contener información sensible.

Las amenazas más comunes son:

- **SPAM**.
- **Phishing**.
- **Malware mediante adjuntos**.


### 7.1. SPAM

El **SPAM** son correos no solicitados enviados normalmente en grandes cantidades.

Características habituales:

- Remitentes desconocidos.
- Mensajes genéricos o poco personalizados.
- Enlaces sospechosos.
- Archivos adjuntos inesperados.
- Promesas exageradas o publicidad intrusiva.

Riesgos:

- Saturación de la bandeja de entrada.
- Posible infección por malware.
- Robo de información si el usuario interactúa con enlaces o adjuntos.

Medidas de protección:

- Usar filtros antispam.
- Configurar SPF, DKIM y DMARC.
- No interactuar con mensajes sospechosos.
- Marcar mensajes como SPAM para entrenar los filtros.


### 7.2. Phishing

El **phishing** es una técnica de ingeniería social que intenta engañar al usuario para que revele información confidencial.

Características habituales:

- Simula proceder de una empresa o institución confiable.
- Usa un tono urgente o alarmista.
- Incluye enlaces a páginas falsas que imitan servicios reales.
- Puede usar direcciones parecidas a las legítimas.

Riesgos:

- Robo de credenciales.
- Acceso no autorizado a cuentas.
- Pérdida económica.
- Robo de identidad.

Medidas de protección:

- Revisar cuidadosamente el remitente.
- No abrir enlaces sospechosos.
- Acceder manualmente a la web oficial.
- Usar autenticación de dos factores.
- Educar a los usuarios para detectar señales de engaño.


### 7.3. Malware mediante adjuntos

El malware puede distribuirse mediante archivos adjuntos en correos electrónicos.

Tipos de archivos especialmente sospechosos:

- Ejecutables: `.exe`, `.bat`, `.scr`.
- Archivos comprimidos: `.zip`, `.rar`, especialmente si no se esperaban.
- Documentos con macros: `.docm`, `.xlsm`.

Riesgos:

- Robo de información.
- Cifrado de archivos mediante ransomware.
- Acceso remoto por parte de un atacante.
- Propagación por la red interna.

Medidas de protección:

- Antivirus actualizado.
- Bloqueo de adjuntos peligrosos en el servidor.
- Deshabilitar macros por defecto.
- No abrir adjuntos inesperados.


### Actividad: detectar señales de phishing

Lee el siguiente mensaje ficticio:

```text
De: seguridad@banco-seguro-verificacion.com
Asunto: URGENTE: Su cuenta será bloqueada hoy

Estimado cliente,
Hemos detectado un problema con su cuenta. Debe verificar sus datos inmediatamente en el siguiente enlace:
http://banco-seguro-login.example/verificar

Si no lo hace en las próximas 2 horas, su cuenta quedará bloqueada.
```

Responde:

1. ¿Qué señales de sospecha aparecen?
2. ¿Qué debería hacer el usuario?
3. ¿Qué medidas técnicas podrían ayudar a reducir este tipo de ataques?


````python
# Pequeño ejemplo: detector muy básico de señales de phishing en un texto
# No sustituye a un filtro real, solo sirve como actividad didáctica.

mensaje = '''
URGENTE: Su cuenta será bloqueada hoy.
Debe verificar sus datos inmediatamente haciendo clic en este enlace.
'''

palabras_sospechosas = ["urgente", "bloqueada", "verificar", "contraseña", "inmediatamente", "premio", "cuenta"]

puntuacion = 0
for palabra in palabras_sospechosas:
    if palabra.lower() in mensaje.lower():
        puntuacion += 1
        print(f"Señal detectada: {palabra}")

print("Puntuación de sospecha:", puntuacion)

if puntuacion >= 3:
    print("Resultado: mensaje potencialmente sospechoso")
else:
    print("Resultado: no hay suficientes señales, revisar manualmente")
````

## 8. Protección del correo electrónico

Para proteger el correo se combinan medidas técnicas y buenas prácticas.

Entre las medidas técnicas más importantes están:

- Filtros antispam.
- Registros SPF.
- Firmas DKIM.
- Políticas DMARC.
- Antivirus y análisis de adjuntos.

### 8.1. Filtros antispam

Los filtros antispam intentan identificar y bloquear mensajes no deseados antes de que lleguen al usuario.

Pueden analizar:

- Remitente.
- Dirección IP de origen.
- Contenido del mensaje.
- Enlaces incluidos.
- Adjuntos.
- Reputación del dominio.
- Cabeceras técnicas.

Tipos habituales:

| Tipo | Funcionamiento |
|---|---|
| Listas negras | Bloquean IPs o dominios conocidos por enviar SPAM. |
| Listas blancas | Permiten remitentes de confianza. |
| Palabras clave | Detectan expresiones frecuentes en SPAM. |
| Bayesianos | Aprenden de mensajes marcados como SPAM o legítimos. |

## 9. SPF: Sender Policy Framework

**SPF** permite indicar qué servidores están autorizados a enviar correos en nombre de un dominio.

Su objetivo principal es reducir la **suplantación de identidad** o *spoofing*.

La configuración SPF se publica como un registro **TXT** en DNS.

Ejemplo:

```text
v=spf1 ip4:192.168.1.1 -all
```

Interpretación:

- `v=spf1`: versión de SPF.
- `ip4:192.168.1.1`: esa IP está autorizada.
- `-all`: cualquier otro servidor debe ser rechazado.


### 9.1. Proceso de verificación SPF

1. Un servidor receptor recibe un correo.
2. Identifica el dominio del remitente.
3. Consulta en DNS el registro SPF del dominio.
4. Compara la IP del servidor que envió el mensaje con las IP autorizadas.
5. Según el resultado, acepta, rechaza o marca el mensaje como sospechoso.

```text
correo@ejemplo.com enviado desde 203.0.113.5
        │
        ▼
Servidor receptor consulta SPF de ejemplo.com
        │
        ▼
¿203.0.113.5 está autorizado?
        │
        ├── Sí → SPF pass
        └── No → fail / softfail / neutral
```


### 9.2. Mecanismos comunes de SPF

| Mecanismo | Significado |
|---|---|
| `all` | Regla final para cualquier servidor no listado. |
| `ip4` | Autoriza una IP o rango IPv4. |
| `ip6` | Autoriza una IP o rango IPv6. |
| `include` | Incluye las reglas SPF de otro dominio. |
| `a` | Autoriza las IP de los registros A del dominio. |
| `mx` | Autoriza los servidores MX del dominio. |
| `ptr` | Autoriza según resolución inversa, aunque se desaconseja en muchas configuraciones modernas. |


### 9.3. Calificadores SPF

| Calificador | Nombre | Significado |
|---|---|---|
| `+` | Pass | Autoriza el envío. Es el valor por defecto si no se indica. |
| `-` | Fail | El envío no está autorizado; normalmente se rechaza. |
| `~` | SoftFail | El envío no está autorizado claramente, pero se acepta y se marca como sospechoso. |
| `?` | Neutral | No se declara una política clara. |


### 9.4. Ejemplos de registros SPF

| Registro | Interpretación |
|---|---|
| `v=spf1 ip4:192.168.1.1 -all` | Solo esa IP está autorizada; el resto se rechaza. |
| `v=spf1 ip4:192.168.1.1 ~all` | Esa IP está autorizada; el resto se marca como sospechoso. |
| `v=spf1 ?all` | No se impone una política estricta. |
| `v=spf1 ip4:203.0.113.1 +ip4:203.0.113.2 ~all` | Autoriza varias IPs y aplica SoftFail al resto. |
| `v=spf1 include:_spf.google.com -all` | Autoriza los servidores incluidos en el SPF de Google y rechaza el resto. |


````python
# Práctica: evaluación simplificada de SPF con ipaddress
# Este ejemplo solo comprueba mecanismos ip4 y el calificador final all.

from ipaddress import ip_address, ip_network

spf = "v=spf1 ip4:203.0.113.0/24 -all"
ip_emisor = "203.0.113.5"

def evaluar_spf_simple(spf_record, ip):
    partes = spf_record.split()
    ip_obj = ip_address(ip)
    politica_final = "neutral"

    for parte in partes:
        if parte == "v=spf1":
            continue
        
        # Calificador explícito o implícito
        calificador = "+"
        mecanismo = parte
        if parte[0] in ["+", "-", "~", "?"]:
            calificador = parte[0]
            mecanismo = parte[1:]
        
        if mecanismo.startswith("ip4:"):
            red = mecanismo.replace("ip4:", "")
            if "/" not in red:
                red += "/32"
            if ip_obj in ip_network(red):
                return "pass" if calificador == "+" else calificador
        
        if mecanismo == "all":
            politica_final = calificador
    
    return {
        "+": "pass",
        "-": "fail",
        "~": "softfail",
        "?": "neutral"
    }.get(politica_final, "neutral")

resultado = evaluar_spf_simple(spf, ip_emisor)
print("Registro SPF:", spf)
print("IP emisora:", ip_emisor)
print("Resultado SPF:", resultado)
````

### 9.5. Beneficios, limitaciones y buenas prácticas de SPF

**Beneficios:**

- Reduce la suplantación de dominios.
- Ayuda a disminuir el SPAM.
- Mejora la reputación del dominio si está bien configurado.

**Limitaciones:**

- No protege el contenido del mensaje.
- Puede fallar en reenvíos legítimos si el servidor intermedio no está autorizado.
- Una configuración incorrecta puede provocar rechazos de correos legítimos.

**Buenas prácticas:**

- Incluir todos los servidores que envían correo en nombre del dominio.
- Usar `-all` solo cuando se esté seguro de la configuración.
- Empezar con `~all` durante fases de prueba.
- Complementar SPF con DKIM y DMARC.
- Revisar periódicamente la configuración.


## 10. DKIM: DomainKeys Identified Mail

**DKIM** añade una firma digital a los correos enviados desde un dominio.

Funcionamiento básico:

1. El servidor emisor firma el mensaje con una clave privada.
2. El servidor receptor obtiene la clave pública desde DNS.
3. El receptor comprueba la firma.
4. Si la firma es válida, se confirma que el mensaje no fue alterado durante el tránsito y que está vinculado al dominio firmante.

Ventaja principal:

- Garantiza la **integridad** del mensaje y ayuda a verificar su autenticidad.


### Ejemplo conceptual de DKIM

Un registro DKIM suele publicarse en DNS como un TXT asociado a un selector.

Ejemplo conceptual:

```text
selector1._domainkey.ejemplo.com TXT "v=DKIM1; k=rsa; p=CLAVE_PUBLICA..."
```

Donde:

- `selector1` identifica qué clave debe usarse.
- `_domainkey` es una parte fija de la estructura DKIM.
- `p=` contiene la clave pública.


## 11. DMARC: Domain-based Message Authentication, Reporting and Conformance

**DMARC** combina SPF y DKIM y permite al propietario de un dominio indicar qué debe hacerse con los mensajes que no superen las comprobaciones.

Puede definir políticas como:

| Política | Significado |
|---|---|
| `p=none` | No rechazar; solo monitorizar. Útil al empezar. |
| `p=quarantine` | Marcar o enviar a cuarentena los mensajes sospechosos. |
| `p=reject` | Rechazar mensajes que no superen las comprobaciones. |

Ejemplo:

```text
_dmarc.ejemplo.com TXT "v=DMARC1; p=quarantine; rua=mailto:dmarc@ejemplo.com"
```

Además, DMARC puede generar informes para que el administrador vea quién intenta enviar correos en nombre de su dominio.


## 12. Comparativa SPF, DKIM y DMARC

| Mecanismo | Qué comprueba | Dónde se configura | Qué aporta |
|---|---|---|---|
| **SPF** | Si la IP emisora está autorizada para enviar por el dominio. | Registro TXT en DNS. | Reduce spoofing basado en servidores no autorizados. |
| **DKIM** | Si el mensaje conserva una firma válida del dominio. | Clave pública en DNS y clave privada en servidor. | Verifica integridad y autenticidad del mensaje. |
| **DMARC** | Si SPF/DKIM se alinean con el dominio y qué política aplicar. | Registro TXT en `_dmarc.dominio`. | Define acciones y genera informes. |
